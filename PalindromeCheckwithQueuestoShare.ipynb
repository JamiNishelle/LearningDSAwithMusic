```python
# Full IsPalindrome() check with queue function

from collections import deque

def is_palindrome(s):
    
    s_alnum_deque = deque()

    for i in range(len(s)):
        if s[i].isalnum():
            s_alnum_deque.append(s[i].lower()) # lowercase the string and add to deque if alphanumeric

    # Now you have a deque of lowercased alphanumeric characters.

    while len(s_alnum_deque) > 1:
        if s_alnum_deque.popleft() == s_alnum_deque.pop():
            pass
        else:
            return False
        
    return True


```

```python
s = ['1', '2','!', 'a', 'B']

is_palindrome(s)
```

```python
s = ['1', '2','!', '2', '1']

is_palindrome(s)
```

```python
print(is_palindrome("A man, a plan, a canal: Panama"))  # Should be True
print(is_palindrome("race a car"))                       # Should be False
print(is_palindrome("a"))                                # Should be True
```

```python
from collections import deque

# Common scales defined as MIDI notes starting from C4 (60)
scales = {
    # 7-note (heptatonic) scales
    "Major (Ionian)":        [60, 62, 64, 65, 67, 69, 71, 72],
    "Dorian":                [60, 62, 63, 65, 67, 69, 70, 72],
    "Phrygian":              [60, 61, 63, 65, 67, 68, 70, 72],
    "Lydian":                [60, 62, 64, 66, 67, 69, 71, 72],
    "Mixolydian":            [60, 62, 64, 65, 67, 69, 70, 72],
    "Natural Minor (Aeolian)":[60, 62, 63, 65, 67, 68, 70, 72],
    "Locrian":               [60, 61, 63, 65, 66, 68, 70, 72],
    "Harmonic Minor":        [60, 62, 63, 65, 67, 68, 71, 72],
    "Melodic Minor":         [60, 62, 63, 65, 67, 69, 71, 72],
    # Pentatonic scales
    "Major Pentatonic":      [60, 62, 64, 67, 69, 72],
    "Minor Pentatonic":      [60, 63, 65, 67, 70, 72],
    "Blues":                 [60, 63, 65, 66, 67, 70, 72],
    # Symmetric scales
    "Whole Tone":            [60, 62, 64, 66, 68, 70, 72],
    "Diminished (whole-half)":[60, 62, 63, 65, 66, 68, 69, 71, 72],
    "Diminished (half-whole)":[60, 61, 63, 64, 66, 67, 69, 70, 72],
    "Chromatic":             [60, 61, 62, 63, 64, 65, 66, 67, 68, 69, 70, 71, 72],
}

def get_intervals(midi_notes):
    """Compute the intervals (semitone differences) between consecutive notes."""
    return [midi_notes[i+1] - midi_notes[i] for i in range(len(midi_notes) - 1)]

def is_interval_palindrome(intervals):
    """Check if an interval list is a palindrome using a deque."""
    dq = deque(intervals)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True

# Check all scales
print(f"{'Scale':<28} {'Intervals':<30} {'Palindrome?'}")
print("-" * 72)
for name, notes in scales.items():
    # Create a new Stream for this melody
    score = stream.Stream()
    
    # Convert MIDI numbers to Note objects and add to stream
    for num in notes:
        n = note.Note(num)
        n.quarterLength = 1.0  # Set duration (1.0 = quarter note)
        score.append(n)
    
    # Write to a MIDI file
    score.write('midi', fp=f'{name}.mid')

    # create the table
    intervals = get_intervals(notes)
    result = is_interval_palindrome(intervals)
    emoji = "✅" if result else "❌"
    print(f"{name:<28} {str(intervals):<30} {emoji}")
```

# Fibonacci Word

The Fibonacci word is a string built by concatenation, just like Fibonacci numbers are built by addition:

- `F(1) = "0"`
- `F(2) = "01"`
- `F(n) = F(n-1) + F(n-2)`

Each word is the **previous word glued to the one before it**.

A known property: if you remove the last two characters, the result is a palindrome!


```python
from collections import deque

# --- Recursive Fibonacci Word ---
def fibonacci_word(n):
    if n == 1:
        return "0"
    if n == 2:
        return "01"
    return fibonacci_word(n - 1) + fibonacci_word(n - 2)  # concatenate, not add!

# --- Palindrome check using a deque (reusing our earlier approach) ---
def is_palindrome_str(s):
    dq = deque(s)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True

# --- Check each Fibonacci word after removing its last 2 characters ---
print(f"{'fib_n':<6} {'F(n)':<45} {'F(n)[:-2]':<43} {'Palindrome?'}")
print("-" * 100)

for fib_n in range(1, 11):
    fib_word = fibonacci_word(fib_n)
    fib_trimmed = fib_word[:-2] if len(fib_word) > 2 else fib_word  # remove last 2 chars

    score = stream.Stream()
    
    # Convert MIDI numbers to Note objects and add to stream
    for num in fib_trimmed:
        n = note.Note(int(num) + 60) 
        n.quarterLength = 1.0  # Set duration (1.0 = quarter note)
        score.append(n)
    
    # Write to a MIDI file
    score.write('midi', fp=f'word_{fib_trimmed}.mid')

    result = is_palindrome_str(fib_trimmed)
    emoji = "✅" if result else "❌"
    # Truncate display for long words
    word_display = fib_word if len(fib_word) <= 43 else fib_word[:40] + "..."
    trimmed_display = fib_trimmed if len(fib_trimmed) <= 41 else fib_trimmed[:38] + "..."
    print(f"{fib_n:<6} {word_display:<45} {trimmed_display:<43} {emoji}")
```

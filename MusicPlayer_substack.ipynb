```python
# Solve Leetcode question: Shuffle an Array https://leetcode.com/problems/shuffle-an-array/
from typing import List
import random


class Solution:
    # Need to store The original array (so reset() can return it)
    # Need to store A copy to shuffle
    def __init__(self, nums: List[int]):
        self.nums = nums
        self.copy = nums.copy()

    def reset(self) -> List[int]:
        return self.nums

    def shuffle(self) -> List[int]:
\
        # Write the Fisher-Yates shuffle on that list
        # Initialize i outside of the while loop
        # Start from the last element (i = len(list) - 1)
        i = len(self.copy) - 1 # i starts as the largest index in list
        while i > 0:
            
            # Pick a random index j between 0 and i
            j = random.randint(0, i) # Return a random integer N such that a <= N <= b.

            # save list[j] first:
            temp = self.copy[j]
            self.copy[j] = self.copy[i] # Swap self.copy[i] and self.copy[j]
            self.copy[i] = temp # put the original value of self.copy[j] back to self.copy[i]

            # Decrease i and repeat
            i -= 1

        return self.copy
```
